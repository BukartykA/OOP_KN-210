## Порівняння простих даних з об'єктами

```python
animals = ["Рибки", "Кіт", "Собака"]
name = ["Неони", "Мурка", "Рекс"]
age = [1, 5, 2]
print(f"В мене є тваринка {animals[1]} яку звати {name[1]} які {age[1]} років")

class MyAnimals:
    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        # Виправлення: використовуємо 'mass'
        self.mass = 0 if mass==None else mass

a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals(animals="Собака", name="Рекс", age=2)

print(f"\nОригінальне ім'я a2: {a2.name}")

# Перейменовуємо об'єкти
a1.name = "Гупії"
a2.name = "Машка"
a3.name = "Хіккі"

print("\n--- Нові імена та атрибути ---")
print(f"a2.name: {a2.name}")
print(f"Я переіменував свою тваринку на {a2.name}, вона важить {a2.mass} кг")
print(f"Моя тваринка {a1.name}, вона важить {a1.mass} кг")
print(f"Всі нові імена: a1: {a1.name}, a2: {a2.name}, a3: {a3.name}")
```

## Методи класу

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
    
    def speak(self):
        if self.animals == "Кіт":
            return f"Моя Тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя Тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя Тваринка {self.animals} не розмовляє"
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"

a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals(animals="Собака", name="Рекс", age=2)

print("\n--- Застосування методів до всіх об'єктів ---")
print(f"Тваринка {a1.animals}:", a1.speak())
print(f"Тваринка {a2.animals}:", a2.speak())
print(f"Тваринка {a3.animals}:", a3.speak())

print(f"\n{a1.animals} живе зі мною {a1.age} років")
print(a1.add_one_year())
print(f"{a1.animals} живе зі мною {a1.age} років")
print(f"Новий вік a1: {a1.age}")

print(f"\n{a2.animals} живе зі мною {a2.age} років")
print(a2.add_one_year())
print(f"Новий вік a2: {a2.age}")
```

## Змінні класу, глобальні змінні над всіма об'єктами

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1

print(f"В мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")

# Створюємо об'єкти в циклі
data = [
    ("Рибки", "Неони", 1),
    ("Рибки", "Гупії", 1),
    ("Рибки", "Соміки", 1),
    ("Кіт", "Мурка", 5, 10),
    ("Собака", "Рекс", 2)
]

all_pets = []
for pet in data:
    if len(pet) == 4:
        all_pets.append(MyAnimals(pet[0], pet[1], pet[2], pet[3]))
    else:
        all_pets.append(MyAnimals(pet[0], pet[1], pet[2]))

a1 = all_pets[0]

print(f"\nВ мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")
print(f"Я доступаюсь до змінної класу з обєкта: {a1.total_animal_types}")

# Створення атрибуту об'єкта з тим же ім'ям, що й змінна класу
a1.total_animal_types = {"Собака"} 

print(f"\n--- Після переприсвоєння a1.total_animal_types ---")
print(f"Я доступаюсь до змінної класу з обєкта (атрибут об'єкта): {a1.total_animal_types}") 
print(f"Я доступаюсь до змінної класу (самого класу): {MyAnimals.total_animal_types}")
```

## Статичні методи та змінна self

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"

a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)

print("\n--- Виклик статичного методу ---")
print(f"Ми кличемо нашу тваринку за допомогою статичного методу з об'єкту a1 |{a1.call_pet()}|")
print(f"Ми кличемо нашу тваринку за допомогою статичного методу з об'єкту a2 |{a2.call_pet()}|")
print(f"Ми кличемо нашу тваринку за допомогою статичного методу з класу |{MyAnimals.call_pet()}|")

print(f"\n--- Виклик методу екземпляра ---")
print(f"Викликаю метод з об'єкту a1 |{a1.add_one_year()}|")
# Виклик методу екземпляра через клас, передаючи екземпляр явно
print(f"Викликаю метод з класу, передаючи a1 |{MyAnimals.add_one_year(a1)}|")
```

## Робота з properties

```python
from datetime import datetime

class MyAnimals:
    max_age = 20
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age
    
    # НОВА PROPERTY: Властивість, що показує вік у людських роках
    @property
    def human_age(self):
        return self.age * 4
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"

a1 = MyAnimals("Кіт", "Мурка", 5)

print(f"\n--- Вивід нової property human_age ---")
print(f"Мій {a1.animals} віком {a1.age} має приблизно {a1.human_age} людських років.")

print(f"\nМій {a1.animals} буде зі мною до {a1.live_untill} року")
a1.add_one_year()
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Після року: Мій {a1.animals} віком {a1.age} має приблизно {a1.human_age} людських років.")
```

## Методи Класу

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)
    
    def __init__(self, animals, name, age) -> None:
        # Валідація імені в конструкторі
        if not name.strip()[0].isupper():
            print(f"Попередження: Ім'я '{name.strip()}' було виправлено на '{name.strip().capitalize()}'")
            self.name = name.strip().capitalize()
        else:
            self.name = name.strip()

        self.animals = animals.strip()
        self.age = int(age) if isinstance(age, str) else age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = [item.strip() for item in csv_data.split(",")]
        return cls(animal, name, age)

# НОВИЙ МЕТОД КЛАСУ ДЛЯ ПЕРЕВІРКИ
    @classmethod
    def check_name_case(cls, name: str):
        return name.strip().capitalize() == name.strip()

a1 = MyAnimals.from_csv("Кіт, мурка, 5") # Ім'я 'мурка' буде виправлено
a2 = MyAnimals("Собака", "Рекс", 2)
a3 = MyAnimals("Папуга", "кеша", 1) # Ім'я 'кеша' буде виправлено

print(f"\nМій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Мій {a3.animals}, звати {a3.name}, має {a3.age} років")

print(f"\nПеревірка імені 'Мурка' через метод класу: {MyAnimals.check_name_case('Мурка')}")
print(f"Перевірка імені 'рекс' через метод класу: {MyAnimals.check_name_case('рекс')}")
print(f"Загальна кількість обєктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")
```

## Приватні атрибути та методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        self._secret_phrase = "Це приватний атрибут, (як я буду кликати свою тваринку)"
        self.__secret_strong = "Цей атрибут точно не можна викликати ззовні!!"
    
    @property
    def secret_phrase(self):
        return f"Ми доступаємось до приватного атрибуту з проперті: {self._secret_phrase}"
    
    @property
    def secret_strong(self):
        return f"Ми доступаємось до приватного атрибуту з проперті: {self.__secret_strong}"

    def _secret_methid(self):
        return "Приватний метод, краще його не рухати, він щось робить потрібне всередині класу"

    def __do_not_touch_it(self):
        return "Це неможливо викликати ззовні"

a1 = MyAnimals("Собака", "Рекс", 2)
print(f"Доступ до _secret_phrase: {a1._secret_phrase}")
a1._secret_phrase = "Я змінив приватний атрибут"  ## Такого робити не треба!
print(f"Після зміни: {a1._secret_phrase}")

# print(a1.__secret_strong) ## Це видасть AttributeError, тому закоментовано!
print(f"Доступ до __secret_strong через Name Mangling: {a1._MyAnimals__secret_strong}") ### Так також робити не можна!

### Виводимо приватний атрибут через проперті
print(f"\nЧерез property: {a1.secret_phrase}")
# a1.secret_phrase = "..." # З проперті так зробити не можна!
print(f"Через property: {a1.secret_strong}")
print(f"Виклик приватного методу: {a1._secret_methid()}")
# a1.__do_not_touch_it() ##Це виклик видасть помилку!
```

## Магічні методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        # Репрезентація для розробника
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        # Репрезентація для користувача
        return f"Це є об'єкт {self.animals} з іменем {self.name}" 

a1 = MyAnimals("Собака", "Рекс", 2)
print(f"Вивід через print(): {a1}") # Викликає __str__
print(f"Вивід через repr(): {repr(a1)}") # Викликає __repr__
print(f"Вивід через str(): {str(a1)}") # Викликає __str__
```

## Арифметичні методи

```python
a = 2
b = 5
c = "C"
print(a+b, a - int(1),  a * b, str(a) + c)
print(a.__add__(b), a.__sub__(int(1)), a.__mul__(b))

class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __sub__(self, obj):
        if isinstance(obj, MyAnimals):
            return f"Мій {self.name} покусав {obj.name}"
        elif isinstance(obj, int):
            # Тут змінюємо вік поточного об'єкта
            self.age += obj
            return f"Мій {self.name} постарів на {obj} років, тепер йому {self.age}"
        return NotImplemented # Для непідтримуваних типів
    
    def __gt__(self, obj):
        if isinstance(obj, MyAnimals):
            if self.age > obj.age:
                return f"Старшим є {self.name}"
            return f"{self.name} не є старшим за {obj.name}"
        return NotImplemented

    def __len__(self) -> int:
        print("Ми зайшли в метод len, та знаходимо довжину Імені тваринки")
        return len(self.name)
            
a = MyAnimals("Собака", "Рекс", 2)
b = MyAnimals("Кіт", "Мурзик", 4)

print(f"\n--- Оператори ---")
print(f"a - b: {a - b}") # Викликає a.__sub__(b)
print(f"a - 1: {a - 1}") # Викликає a.__sub__(1)
# print(f"a - float(1): {a - float(1)}") # Виклик з float(1) не оброблено, поверне TypeError
print(f"a > b: {a > b}") # Викликає a.__gt__(b)
print(f"b > a: {b > a}") # Викликає b.__gt__(a)
print(f"len(a): {len(a)}") # Викликає a.__len__()
```
